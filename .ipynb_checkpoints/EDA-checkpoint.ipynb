## Solución ejercicio 3

```python
x = vdf.loc[ : , ['item', 'variety', 'Order Date Year', 'date', 'price'] ].drop_duplicates()
```

```python
x
```

```python
x = x.sort_values(['item', 'variety', 'Order Date Year', 'date', 'price'])
```

```python
x.groupby(['item', 'variety', 'Order Date Year', 'price'], as_index=False)['date'].shift(1)
```

```python
x['FechaPrevia'] = x.groupby(['item', 'variety', 'Order Date Year', 'price'], as_index=False)['date'].shift(1)
x
```

```python
x = x.loc[ ~pd.isna(x['FechaPrevia']) , : ]
```

```python
x.insert(0, 'DiasTranscurridosDesdeUltimaCompra', (x['date'] - x['FechaPrevia']).dt.days )
```

```python
x
```

```python
x.groupby(['item', 'variety', 'Order Date Year', 'price'], as_index = False).agg( { 'DiasTranscurridosDesdeUltimaCompra': ['mean', 'median'] })
```

### Ejercicio seis: muestre por category,item,variety que tanto % ha aumentado su precio

Ejemplo: si para las manzanas gala el primer precio que se registro fue 2.00 y el ultimo precio que se registro es 4.00 este aumento un 100%

```python
pd.DataFrame({
    "category": ["fruta","fruta"]
    ,"item" : ['Manzana','Manzana']
    , "variety" : ['Gala','egremont_russet']
    , "percentage-change-between-first-and-last-price" : [200,150]       
})
```

## Solución ejercicio 6

```python
vdf['percentage-change-between-first-and-last-price'] = (vdf['price'] / 
                      vdf['price'].sum()) * 100
```

```python
vdf_sin_repetidos = vdf.loc[:, ['category', 'item','variety', 'percentage-change-between-first-and-last-price'] ].drop_duplicates()
```

```python
vdf_sin_repetidos = vdf_sin_repetidos.sort_values(['category', 'item','variety', 'percentage-change-between-first-and-last-price'])
```

```python
vdf_sin_repetidos.groupby(['category', 'item', 'variety'], as_index = False).agg({ 'percentage-change-between-first-and-last-price': ['mean', 'median']})
```
